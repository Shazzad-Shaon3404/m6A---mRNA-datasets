```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
dataset = pd.read_csv('/content/drive/MyDrive/a new RNA m6A/dataset/main data for merged/100-XGB-Features.csv')

X = dataset.drop('Target', axis=1)
y = dataset['Target']

model = XGBClassifier() # xgb for -----feature extraction---------------------------------------------------------------------------------------------
model.fit(X, y)

importance_scores = model.feature_importances_


num_features = len(importance_scores)


colors = sns.color_palette("husl", num_features)


plt.figure(figsize=(14, 8))
plt.bar(range(num_features), importance_scores, color=colors)
plt.xlabel('Feature')
plt.ylabel('Importance Score')
plt.title('Feature Importance Scores')

plt.savefig('FT.png', dpi=800)
plt.savefig('ROC Curves.pdf', dpi=800)
plt.show()
```

```python

feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': importance_scores})


top_20_features = feature_importance_df.sort_values(by='Importance', ascending=False).head(20)
num_features = len(top_20_features)
colors = sns.color_palette("husl", num_features) #husl color ------------


plt.figure(figsize=(14, 8))
plt.bar(range(len(top_20_features)), top_20_features['Importance'], tick_label=top_20_features['Feature'], color=colors)
plt.xlabel('Feature')
plt.ylabel('Importance Score')
plt.title('Top 20 Feature Importance Scores')
plt.savefig('FT20.png', dpi=800)
plt.savefig('ROC Curves.pdf', dpi=800)
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

sensitivity_aminer, specificity_aminer = 0.806, 0.861
sensitivity_mpred, specificity_mpred = 0.803, 0.849
sensitivity_dlm, specificity_dlm = 0.796, 0.858
sensitivity_cat, specificity_cat = 0.842, 0.9109

plt.figure(figsize=(6, 5))

fpr_smooth = np.linspace(0, 1, 100)
interp_tpr_aminer = interp1d([0, 1 - specificity_aminer, 1], [0, sensitivity_aminer, 1])
interp_tpr_mpred = interp1d([0, 1 - specificity_mpred, 1], [0, sensitivity_mpred, 1])
interp_tpr_dlm = interp1d([0, 1 - specificity_dlm, 1], [0, sensitivity_dlm, 1])
interp_tpr_cat = interp1d([0, 1 - specificity_cat, 1], [0, sensitivity_cat, 1])
plt.plot(fpr_smooth, interp_tpr_aminer(fpr_smooth), label='m6Aminer', color='green')
plt.plot(fpr_smooth, interp_tpr_mpred(fpr_smooth), label='m6AmPred ', color='cyan')
plt.plot(fpr_smooth, interp_tpr_dlm(fpr_smooth), label='DLm6Am', color='orange')
plt.plot(fpr_smooth, interp_tpr_cat(fpr_smooth), label='m6A-DLcat', color='blue')
plt.scatter(1 - specificity_aminer, sensitivity_aminer, marker='o', color='green')
plt.scatter(1 - specificity_mpred, sensitivity_mpred, marker='o', color='cyan')
plt.scatter(1 - specificity_dlm, sensitivity_dlm, marker='o', color='orange')
plt.scatter(1 - specificity_cat, sensitivity_cat, marker='o', color='blue')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Specificity')
plt.ylabel(' Sensitivity')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.legend()
plt.savefig('comp anti.png', dpi=800)
plt.savefig('ROC Curves.pdf', dpi=800)
plt.show()
```
